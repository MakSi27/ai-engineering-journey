```python
# Perform matrix multiplication manually — using a loop
matrix1 = [[1,2],
[3,4]]
matrix2 = [[5,6],
[7,8]]

# initialy resultant matrix will be a zero matrix with same dimentions
result = [[0,0],
[0,0]]

for i in range(len(matrix1)):
    for j in range(len(matrix2[0])):
        for k in range(len(matrix2)):
            result[i][j] += matrix1[i][k]*matrix2[k][j]
for row in result:
    print(row)
```

```python
matrix = np.array([[1,2,3],[4,5,6]])
print("shape of matrix befor transpose:", matrix.shape)
matrix = matrix.T
print("shape of matrix after transpose:" , matrix.shape)
```

```python
matrix1 = np.array([[1,2,3],[1,2,3]])
matrix2 = np.array([[4,5],[4,5],[4,5]])
result = matrix1 @ matrix2
print(result.shape)
result2 = matrix2 @ matrix1
print(result2.shape)
```

```python
# Example of error of mismatch
matrix1 = np.array([[1,2,3],[1,2,3]])
matrix2 = np.array([[4,5,3],[4,5,3],[4,5,3]])
result = matrix1 @ matrix2
print(result.shape)
result2 = matrix2 @ matrix1
print(result2.shape)
```

```python
matrix = np.array([[3,4,5],[6,7,8],[2,5,9]])
determinant =np.linalg.det(matrix)
print(determinant)
inverse = np.linalg.inv(matrix)
print(inverse)
```
