```python
import numpy as np
```

```python
# Create vector
data = np.arange(1,10)
vector1,vector2,vector3 = data.reshape(3,3)
print("vector1:", vector1)
print("vector2:", vector2)
print("vector3:", vector3)
```

```python
# Create vector
data = np.arange(1,10)
vector1 , vector2 , vector3 = np.split(data,3)
print("vector1:", vector1)
print("vector2:", vector2)
print("vector3:", vector3)
```

```python
# Create 3 matrices
mat_data = np.arange(1,13).reshape(3,2,2)

matrix1,matrix2,matrix3 = mat_data
print("matrix1:",matrix1)
print("matrix2:",matrix2)
print("matrix3:",matrix3)
```

```python
# Perform matrix multiplication manually — using a loop
matrix1 = [[1,2],
[3,4]]
matrix2 = [[5,6],
[7,8]]

# initialy resultant matrix will be a zero matrix with same dimentions
result = [[0,0],
[0,0]]

for i in range(len(matrix1)):
    for j in range(len(matrix2[0])):
        for k in range(len(matrix2)):
            result[i][j] += matrix1[i][k]*matrix2[k][j]
for row in result:
    print(row)
```
